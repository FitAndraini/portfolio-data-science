## **`Problem`**
Teks yang akan dianalisis tidak memiliki konteks yang jelas, sulit untuk menentukan sentimen yang tepat.

- Public Opinion (mengetahui apakah masyarakat mendukung atau menentang salah satu pihak yang terlibat dalam konflik, sehingga dapat membantu pemerintah atau organisasi terkait dalam memahami tingkat dukungan masyarakat terhadap kebijakan atau tindakan yang diambil)

# Preprocessing

## **`Data Cleaning`**

```python
import re
import string 


DataCleaning = []
for row in dataset:
    print(row)
    kalimat = row
    kalimat = re.sub(r'pic.twitter\S+','',kalimat) #hapus pic twitter
    kalimat2 = re.sub(r'http\S+','',kalimat) #hapus http
    kalimat3= re.sub(r'@\S+','',kalimat2) #hapus @username
    kalimat4 = re.sub(r'#\S+', '', kalimat3) #hapus hastag
    kalimat5 = kalimat4.translate(str.maketrans('','',string.punctuation)).lower()
    kalimat6 = kalimat5.rstrip('\r\n')
    kalimat7 = re.sub(r'[\n\r]+',' ',kalimat6)
    kalimat8 = kalimat7.replace('â€¦','')
    kalimat9 = re.sub (r'[0-9]+','', kalimat8)
    kalimat10 = kalimat9.replace('   ',' ')
    kalimat11 = kalimat10.replace('  ',' ')
    kalimat12 = kalimat11.replace('-',' ')
    kalimat13 = kalimat12.replace(' - ','')
    kalimat14 = ''.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "",kalimat13))
    
    
    print(kalimat14)
    print()
    
    DataCleaning.append(kalimat14)
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer

Tfidf_vector = TfidfVectorizer (max_features=5000)
Tfidf_vector.fit(x)
Train_Tfidf = Tfidf_vector.transform(x_train)
Test_Tfidf = Tfidf_vector.transform(x_test)

print(Tfidf_vector.idf_)
```

# Klasifikasi SVM

```python
from sklearn.feature_extraction.text import CountVectorizer #convert test comment into a numric vector
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC #support vector classification
from sklearn.pipeline import Pipeline #pipeline to implement steps in series
from sklearn import svm 

#classifier SVM
SVM =svm.SVC(C=1.0, kernel='linear') #c = parameter/konstanta
SVM.fit(Train_Tfidf, y_train)
#score = cross_val_score(SVM, x, y, cv=5)

#predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_Tfidf)
print(predictions_SVM)
#print(score)
print("\n")
```

# Evaluasi Model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

akurasi = accuracy_score(predictions_SVM, y_test)
presisi = precision_score(predictions_SVM, y_test, average='macro')
recall = recall_score(predictions_SVM, y_test, average='macro')

print ("akurasi : ", akurasi * 100)
print ("presisi : ", presisi * 100)
print ("recall : ", recall * 100)
```
